```python
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
```

```python
x = np.linspace(-10,10,100)
```

```python
sigma_1 = 0.5
```

```python
sigma_2 = 2.5
```

```python
G1 = np.exp(-1/2*(x/sigma_1)**2)/np.sqrt(2*pi*sigma_1**2)
```

```python
G2 = np.exp(-1/2*(x/sigma_2)**2)/np.sqrt(2*pi*sigma_2**2)
```

```python
np.trapz(G1,x)
```

```python
np.trapz(G2,x)
```

```python
plt.plot(x,G1)
plt.plot(x,G2)
```

```python
plt.plot(x/sigma_1,G1*sigma_1)
plt.plot(x/sigma_2,G2*sigma_2)
```
