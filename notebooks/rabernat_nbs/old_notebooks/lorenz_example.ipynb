# Interactive Visualization of Lorenz Equations

Modeled after [this post](http://jakevdp.github.io/blog/2013/02/16/animating-the-lorentz-system-in-3d/) by Jake VanderPlas

```python
import numpy as np
from scipy import integrate

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import cnames
from IPython.html.widgets import interact, interactive, fixed
from IPython.display import display

%matplotlib inline
```

```python
N_trajectories = 20

def lorenz_deriv((x, y, z), t0, sigma=10., beta=8./3, rho=28.0):
    """Compute the time-derivative of a Lorentz system."""
    return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]


# Choose random starting points, uniformly distributed from -15 to 15
np.random.seed(1)
x0 = -15 + 30 * np.random.random((N_trajectories, 3))

# Solve for the trajectories
t = np.linspace(0, 4, 1000)
x_t = np.asarray([integrate.odeint(lorenz_deriv, x0i, t)
                  for x0i in x0])
```

```python
# animation function.  This will be called sequentially with the frame number
def animate(time=0, angle=10, azimuth=30):

    # Set up figure & 3D axis for animation
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_axes([0, 0, 1, 1], projection='3d')
    ax.axis('off')

    # choose a different color for each trajectory
    colors = plt.cm.jet(np.linspace(0, 1, N_trajectories))

    lines = [l for c in colors for l in ax.plot([], [], [], '-', c=c)]
    pts = [pt for c in colors for pt in ax.plot([], [], [], 'o', c=c)]

    # prepare the axes limits
    ax.set_xlim((-25, 25))
    ax.set_ylim((-35, 35))
    ax.set_zlim((5, 55))

    # we'll step two time-steps per frame.  This leads to nice results.
    #i = (2 * i) % x_t.shape[1]
    i = time
    
    for line, pt, xi in zip(lines, pts, x_t):
        x, y, z = xi[:i].T
        line.set_data(x, y)
        line.set_3d_properties(z)

        pt.set_data(x[-1:], y[-1:])
        pt.set_3d_properties(z[-1:])

    ax.view_init(azimuth, angle)
    fig.canvas.draw()
    plt.show()

#init()

#animate(time=100.,view_angle=10)
w = interactive(animate, azimuth=(-90,90,15.), angle=(0.,360.,15.), time=(1,1000,5))
display(w)

```
